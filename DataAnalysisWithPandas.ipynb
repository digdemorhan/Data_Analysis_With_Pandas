```python
#Age sütunundaki değerlerin ortalamasının alınıp NaN değerler yerine yazılması
age_mean = df['Age'].mean() #mod, medyan, standart sapma vb. de alınabilir
df['Age'].fillna(age_mean, inplace=True)
df['Age'] = df['Age'].round().astype(int)
print(df.to_string())
```

```python
#çok fazla NaN değer içeren Cabin sütununun kaldırılması
df.drop(columns=['Cabin'], inplace=True)

print("Cabin sütunu kaldırıldı.")
print(df.columns)
```

```python
#Embarked sütunundaki NaN değerlerin sütundaki herhangi bir değerle manuel doldurulması
df['Embarked'].fillna('S',inplace=True)
print("Embarked sütunu güncellendi.")
print(df['Embarked'])
```

```python
#Datasette NaN değer kontrolü
nan_count = df.isnull().sum()
print(nan_count[nan_count > 0])
print()
print(df.isnull().sum())
```

```python
#Tekrarlayan satırların kontrolü
duplicates = df[df.duplicated()]
if not duplicates.empty:
    print("Tekrarlayan satırlar var:")
    print(duplicates)
    # Tekrarlayan satırları silme işlemi
    df.drop_duplicates(inplace=True)
    print("Tekrarlayan satırlar silindi.")
else:
    print("Tekrarlayan satır yok.")
```

```python
#Sayısal sütunlar arası korelasyon
numeric_df = df.select_dtypes(include=['number'])
correlation_matrix = numeric_df.corr()
print(correlation_matrix)
```

```python
#Korelasyon tablosunun oluşturulması
correlation_matrix = numeric_df.corr().stack()
correlation_matrix = correlation_matrix[correlation_matrix.index.get_level_values(0) != correlation_matrix.index.get_level_values(1)]

#Korelasyonları en yüksekten en düşüğe sıralanıp ilk 5 en yüksek ve ilk 5 en düşük değerin alınması
highest_correlations = correlation_matrix.sort_values(ascending=False).head(5)
lowest_correlations = correlation_matrix.sort_values().head(5)

print("En yüksek korelasyonlar:")
print(highest_correlations)

print("\nEn düşük korelasyonlar:")
print(lowest_correlations)
```

```python
#Sütunlar üzerinde istatistiksel işlemler
mean_age = df['Age'].mean()
print("Yaş ortalaması:",mean_age)
median_age = df['Age'].median()
print("Yaş medyanı:",median_age)
std_age = df['Age'].std()
print("Yaş standart sapması:",std_age)
mode_age = df['Age'].mode()[0]
print("Yaş modu:",mode_age)
max_age = df['Age'].max()
print("Yaş maksimum değeri:",max_age)
min_age = df['Age'].loc[df['Age'] > 0].min()
print("Yaş minimum değeri:",min_age)
print()
mean_fare = df['Fare'].mean()
print("Ücret ortalaması:",mean_fare)
median_fare = df['Fare'].median()
print("Ücret medyanı:",median_fare)
std_fare = df['Fare'].std()
print("Ücret standart sapması:",std_fare)
mode_fare = df['Fare'].mode()[0]
print("Ücret modu:",mode_fare)
max_fare = df['Fare'].max()
print("Ücret maksimum değeri:",max_fare)
min_fare = df['Fare'].loc[df['Fare'] > 0].min()
print("Ücret minimum değeri:",min_fare)
```

```python
#Belirli sütunlardaki en büyük ve en küçük değerleri döndürme
top_ages = df[df['Age'] > 0]['Age'].nlargest(5)
print("En büyük beş yaş:\n",top_ages)
print()
min_ages = df[df['Age'] > 0]['Age'].nsmallest(5)
print("En küçük beş yaş:\n",min_ages)
print()
top_fares = df[df['Fare'] > 0]['Fare'].nlargest(5)
print("En büyük beş ücret:\n",top_fares)
print()
min_fares = df[df['Fare'] > 0]['Fare'].nsmallest(5)
print("En küçük beş ücret:\n",min_fares)
```

```python
#Embarked sütunundaki değerlerin frekansı
embarked_counts = df['Embarked'].value_counts()
print(embarked_counts)
print()
#Sex sütunundaki değerlerin frekansı
sex_counts = df['Sex'].value_counts()
print(sex_counts)
print()
#Survived sütunundaki değerlerin frekansı
survived_counts = df['Survived'].value_counts()
print(survived_counts)
print()
#Pclass sütunundaki değerlerin frekansı
pclass_counts = df['Pclass'].value_counts()
print(pclass_counts)
print()
#Sibsp sütunundaki değerlerin frekansı
sibsbp_counts = df['SibSp'].value_counts()
print(sibsbp_counts)
print()
#Parch sütunundaki değerlerin frekansı
parch_counts = df['Parch'].value_counts()
print(parch_counts)
```

```python
#Gruplama işlemi
survival_rates = df.groupby(['Sex', 'Pclass'])['Survived'].mean().unstack()
print(survival_rates)
```

***Yukarıdaki kodda bulunan gruplama işlemi, cinsiyet ve yolcu sınıfının hayatta kalma oranına etkisini göstermektedir. Analiz sonucuna bakıldığında kadınların ve 1.sınıf yolcuların hayatta kalma oranı daha yüksektir.***

```python
#Veri görselleştirme
import matplotlib.pyplot as plt
import numpy as np

survival_rates = df.groupby(['Sex', 'Pclass'])['Survived'].mean().unstack()

survival_rates.plot(kind='bar', figsize=(10, 6), color=['#1f77b4', '#ff7f0e', '#2ca02c'], edgecolor='black')

plt.title('Cinsiyete ve Yolcu Sınıfına Göre Hayatta Kalma Oranları', fontsize=16)
plt.xlabel('Cinsiyet', fontsize=14)
plt.ylabel('Hayatta Kalma Oranı', fontsize=14)
plt.xticks(rotation=0)
plt.ylim(0, 1)

plt.legend(title='Yolcu Sınıfı', labels=['1. Sınıf', '2. Sınıf', '3. Sınıf'])

plt.tight_layout()
plt.show()
```

```python
#Gradio Arayüzü
import gradio as gr
import pandas as pd
from io import StringIO

def analyze_data(file):
    #CSV dosyasını okuma
    df = pd.read_csv(file.name)

    #Dataset hakkında genel bilgi
    buffer = StringIO()
    df.info(buf=buffer)
    info_output = buffer.getvalue()

    #Datasetin istatistiksel özeti
    describe_output = df.describe().to_string()

    #Eksik veri sayısını kontrol etme
    missing_values = df.isnull().sum()
    missing_report = f"Eksik veri sayısı:\n{missing_values[missing_values > 0]}\n\n"
    missing_report += f"Toplam eksik veri içeren satır sayısı: {df.isnull().any(axis=1).sum()}\n"
    missing_report += f"Toplam eksik veri içeren sütun sayısı: {df.isnull().any(axis=0).sum()}\n"
    missing_report += f"En çok eksik verinin bulunduğu sütun: {missing_values.idxmax()}\n"
    missing_report += f"En çok eksik verinin bulunduğu satır: {missing_values.idxmax()}\n"

    #Eksik verilerin doldurulması
    df['Age'].fillna(df['Age'].mean(), inplace=True)
    df.drop(columns=['Cabin'], inplace=True)
    df['Embarked'].fillna('S', inplace=True)

    #Datasette NaN değer kontrolü
    nan_count = df.isnull().sum()
    nan_report = f"NaN değer sayıları:\n{nan_count[nan_count > 0]}\n\n"
    nan_report += f"Toplam NaN değer sayısı:\n{nan_count.sum()}\n"

    #Tekrarlayan satırların kontrolü ve varsa silinmesi
    duplicated_rows = df.duplicated()
    duplicate_report = f"Tekrarlayan satır var mı: {duplicated_rows.any()}\n"
    duplicate_report += f"Tekrarlayan satır sayısı: {duplicated_rows.sum()}\n"
    df.drop_duplicates(inplace=True)

    #Sayısal sütunlar arası korelasyon, tablo oluşturulması ve en yüksek/en düşük değerlerin alınması
    numeric_df = df.select_dtypes(include=['number'])
    correlation_matrix = numeric_df.corr()
    correlation_output = correlation_matrix.to_string()
    correlation_report = correlation_matrix.stack()
    correlation_report = correlation_report[correlation_report.index.get_level_values(0) != correlation_report.index.get_level_values(1)]
    highest_correlations = correlation_report.sort_values(ascending=False).head(5)
    lowest_correlations = correlation_report.sort_values().head(5)
    correlation_output += "\n\nEn yüksek korelasyonlar:\n" + str(highest_correlations) + "\n\n"
    correlation_output += "En düşük korelasyonlar:\n" + str(lowest_correlations)

    #Sütunlar üzerinde istatistiksel işlemler
    age_stats = {
        "Yaş Ortalaması": df['Age'].mean(),
        "Yaş Medyanı": df['Age'].median(),
        "Yaş Standart Sapması": df['Age'].std(),
        "Yaş Modu": df['Age'].mode()[0],
        "Yaş Maksimum Değeri": df['Age'].max(),
        "Yaş Minimum Değeri": df['Age'].loc[df['Age'] > 0].min()
    }
    fare_stats = {
        "Ücret Ortalaması": df['Fare'].mean(),
        "Ücret Medyanı": df['Fare'].median(),
        "Ücret Standart Sapması": df['Fare'].std(),
        "Ücret Modu": df['Fare'].mode()[0],
        "Ücret Maksimum Değeri": df['Fare'].max(),
        "Ücret Minimum Değeri": df['Fare'].loc[df['Fare'] > 0].min()
    }
    stats_output = "Yaş İstatistikleri:\n" + str(age_stats) + "\n\n" + "Ücret İstatistikleri:\n" + str(fare_stats)

    #Belirli sütunlardaki en büyük ve en küçük değerleri döndürme
    top_ages = df[df['Age'] > 0]['Age'].nlargest(5)
    min_ages = df[df['Age'] > 0]['Age'].nsmallest(5)
    top_fares = df[df['Fare'] > 0]['Fare'].nlargest(5)
    min_fares = df[df['Fare'] > 0]['Fare'].nsmallest(5)
    extremes_output = f"En büyük beş yaş:\n{top_ages}\n\nEn küçük beş yaş:\n{min_ages}\n\n"
    extremes_output += f"En büyük beş ücret:\n{top_fares}\n\nEn küçük beş ücret:\n{min_fares}"

     #Sütun frekansları
    embarked_counts = df['Embarked'].value_counts()
    sex_counts = df['Sex'].value_counts()
    survived_counts = df['Survived'].value_counts()
    pclass_counts = df['Pclass'].value_counts()
    sibsp_counts = df['SibSp'].value_counts()
    parch_counts = df['Parch'].value_counts()
    frequency_output = (
        f"Embarked frekansı:\n{embarked_counts}\n\n"
        f"Sex frekansı:\n{sex_counts}\n\n"
        f"Survived frekansı:\n{survived_counts}\n\n"
        f"Pclass frekansı:\n{pclass_counts}\n\n"
        f"SibSp frekansı:\n{sibsp_counts}\n\n"
        f"Parch frekansı:\n{parch_counts}\n"
    )

    #Gruplama işlemi
    survival_rates = df.groupby(['Sex', 'Pclass'])['Survived'].mean().unstack()
    survival_output = survival_rates.to_string()

    #Veri görselleştirme
    plt.figure(figsize=(10, 6))
    survival_rates.plot(kind='bar', color=['#1f77b4', '#ff7f0e', '#2ca02c'], edgecolor='black')
    plt.title('Cinsiyete ve Yolcu Sınıfına Göre Hayatta Kalma Oranları', fontsize=16)
    plt.xlabel('Cinsiyet', fontsize=14)
    plt.ylabel('Hayatta Kalma Oranı', fontsize=14)
    plt.xticks(rotation=0)
    plt.ylim(0, 1)
    plt.legend(title='Yolcu Sınıfı', labels=['1. Sınıf', '2. Sınıf', '3. Sınıf'])
    plt.tight_layout()
    plt.savefig('survival_rates.png')
    plt.close()

    #Doldurulmuş veri çıktısı
    filled_report = df

    return info_output, describe_output, missing_report, filled_report, nan_report, duplicate_report, correlation_output, stats_output, extremes_output, frequency_output, survival_output, "survival_rates.png"

gr.Interface(
    fn=analyze_data,
    inputs=gr.File(label="Load CSV"),
    outputs=[gr.Textbox(label="Dataset Hakkında Genel Bilgi", lines=10),
             gr.Textbox(label="Datasetin İstatistiksel Özeti", lines=10),
             gr.Textbox(label="Eksik Veri Raporu", lines=10),
             gr.Dataframe(label="Güncel Dataset"),
             gr.Textbox(label="NaN Değer Kontrolü", lines=10),
             gr.Textbox(label="Tekrarlayan Satır Kontrolü", lines=10),
             gr.Textbox(label="Korelasyon Analizi", lines=10),
             gr.Textbox(label="İstatistiksel Rapor", lines=10),
             gr.Textbox(label="En Büyük ve En Küçük Değerler", lines=10),
             gr.Textbox(label="Frekans Raporu", lines=10),
             gr.Textbox(label="Gruplama Raporu", lines=10),
             gr.Image(label="Hayatta Kalma Oranı Grafiği")
             ],
    title="Data Analysis With Pandas",
    description="Load CSV file."
).launch()
```
